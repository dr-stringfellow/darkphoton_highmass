```python
def transform(gq,gdm=1.0,iRatio=3.,iRelic=False):
    #eps=gq/0.3
    #eps=gq
    out=(2/np.sqrt(1.-0.23))*(128./4./3.14)*gq*gq*np.power(1./iRatio,4.)*gdm*gdm/np.pi/4.
    out=np.where(iRelic,out/gdm/gdm,out)
    return out

def scale(eps,gdm=1.0,iRatio=3.,iRelic=False):
    out=eps*eps*np.power(1./iRatio,4.)*gdm*gdm/np.pi/4.
    out=np.where(iRelic,out/gdm/gdm,out)
    return out
```
